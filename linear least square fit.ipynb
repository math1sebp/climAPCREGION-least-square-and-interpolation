```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#used dataset: https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/global/time-series/globe/land_ocean/1/4/1850-2023/data.csv

# Read the data from a text file, skipping the header row
data = np.loadtxt('monthly_global_mean_temp.txt', delimiter = ',', skiprows = 1)

# Extract the Year and Value columns
years  = data[:, 0]
values = data[:, 1]




# Define the linear function
def linear_func(x, a, b):
    return a * x + b

# Function to perform linear least squares fitting
def perform_linear_least_squares_fit(x, y):
    params, _ = curve_fit(linear_func, x, y)
    return linear_func(x, *params), params

# Perform the linear least squares fit
fit_curve, fit_params = perform_linear_least_squares_fit(years, values)

# Plot the original data and the fitted line
plt.plot(years, values, label = 'Original Data')
plt.plot(years, fit_curve, label = 'Linear Fit')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Linear Least Squares Fit')
plt.legend()
plt.grid(True)
plt.show()
```
